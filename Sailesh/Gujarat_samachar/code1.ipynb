```python
with open('./gujaratsamachar_url.json', 'r') as f:

    url_dict = json.load(f)

try:
    text_df = pd.read_csv('./gujaratsamachar_text1.csv')

except FileNotFoundError:
    text_df = pd.DataFrame(columns=['Link','Text'])

for idx, url in enumerate(url_dict.keys()):

    if url_dict[url] == 2:
        continue

    url_ = "https://www.gujaratsamachar.com" + url

    text = []
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = re.get(url_, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    x = soup.find_all('p')
    for p in x:  
        text.append(p.get_text())
    
    text_df.loc[len(text_df),:] = [url_, text]
    url_dict[url] = 2

    if idx%100 == 0:
        print(idx)
        with open('./gujaratsamachar_url.json', 'w') as f:

            json.dump(url_dict, f, indent=4)

        text_df.to_csv("./gujaratsamachar_text1.csv", index=False)
```
